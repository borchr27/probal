```python
# load data from csv
# turn off user warnings from seaborn
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
labels = ['Atm', 'Beauty', 'Bills And Household', 'Car' ,'Children' ,'Consumer Goods',
                'Culture', 'Digital Services' ,'Drugstore', 'Electronics', 'Fashion',
                'Financial Services', 'Food And Drink' ,'Freetime', 'Groceries', 'Health',
                'House And Garden', 'Investments', 'Pets' ,'Professional Services',
                'Shopping Online', 'Sport', 'Travel']
xpal_filtered_df = pd.read_csv('samples_all_data_filtered_xpal_0.25_cosine_mean_415_7.csv')
xpal_original_df = pd.read_csv('samples_tfidf_data_xpal_0.25_cosine_mean_415_183007.csv')
qbc_filtered_df = pd.read_csv('samples_all_data_filtered_qbc_0.25_cosine_mean_415_7.csv')
qbc_original_df = pd.read_csv('samples_tfidf_data_qbc_0.25_cosine_mean_415_183007.csv')
sns.distplot(xpal_filtered_df['labels'], hist=True, label='xPAL filtered data', bins=23,)
sns.distplot(xpal_original_df['labels'], hist=True, label='xPAL original data', bins=23,)
# sns.distplot(qbc_filtered_df['labels'], hist=True, label='QBC filtered data', bins=23,)
# sns.distplot(qbc_original_df['labels'], hist=True, label='QBC original data', bins=23,)
# show only y axis grid
plt.grid(axis='y')
# use list for labels
plt.xticks(range(0, 23), labels, rotation=90, fontsize=8)
plt.xlabel('Labels', fontsize=12)
plt.legend()
plt.show()
```

```python

import numpy as np
# create swarmplot using seaborn 
# add a column to the xpal filtered data frame with the label 'xPAL filtered data'
xpal_filtered_df['type'] = 'xPAL filtered data'
# add a column to the xpal original data frame with the label 'xPAL original data'
xpal_original_df['type'] = 'xPAL original data'
# combine and plot the xpal data frames
xpal_df = pd.concat([xpal_filtered_df, xpal_original_df])
plt.scatter(xpal_filtered_df.index, xpal_filtered_df['labels'], marker='o', s=5, label='xPAL filtered data')
plt.grid(axis='y')
plt.yticks(range(0, 23), labels, fontsize=8)
#show legend outside of plot
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# NEED TO RUN %7 to get this plot for QBC RANDOM and xPAL
sns.swarmplot(x='labels', y='index', data=xpal_df, size=3, hue='type')
plt.legend(bbox_to_anchor=(0.3, 1.15), loc=2, borderaxespad=0., ncol=1)

plt.grid(axis='y')
plt.xticks(range(0, 23), labels, fontsize=8, rotation=90)
plt.xlabel('Labels', fontsize=12)
plt.ylabel('Index', fontsize=12)
plt.show()
```
