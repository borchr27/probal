```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
warnings.filterwarnings('ignore')
labels = ['Atm', 'Beauty', 'Bills And Household', 'Car' ,'Children' ,'Consumer Goods',
                'Culture', 'Digital Services' ,'Drugstore', 'Electronics', 'Fashion',
                'Financial Services', 'Food And Drink' ,'Freetime', 'Groceries', 'Health',
                'House And Garden', 'Investments', 'Pets' ,'Professional Services',
                'Shopping Online', 'Sport', 'Travel']
xpal_df = pd.read_csv('samples_text_data_all_xpal_0.25_cosine_mean_300_7.csv')
random_df = pd.read_csv('samples_text_data_all_random_0.25_cosine_mean_300_7.csv')
alce_df = pd.read_csv('samples_text_data_all_alce_0.25_cosine_mean_300_7.csv')
```

```python
# add a column to the xpal filtered data frame with the label 'xPAL'
xpal_df['type'] = 'xPAL data'
# add a column to the xpal original data frame with the label 'random'
random_df['type'] = 'random data'
# add a column to the xpal original data frame with the label 'alce'
alce_df['type'] = 'ALCE data'
# combine and plot the xpal data frames
df = pd.concat([xpal_df, alce_df])


sns.swarmplot(x='labels', y='index', data=df, size=3, hue='type')
plt.legend(bbox_to_anchor=(0.3, 1.15), loc=2, borderaxespad=0., ncol=1)

plt.grid(axis='y')
plt.xticks(range(0, 23), labels, fontsize=8, rotation=90)
plt.xlabel('Labels', fontsize=12)
plt.ylabel('Index', fontsize=12)
plt.show()
```
